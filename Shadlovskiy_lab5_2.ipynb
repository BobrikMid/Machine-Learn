```python
import pandas as pd
file_path = "C:/Users/faika/Downloads/archive (2)/StudentData.csv"
df = pd.read_csv(file_path)
print(df.head())
```

```python
print(df.shape)
print(df.dtypes)
```

```python
print(df.isnull().sum())
df.fillna(df.mean(), inplace=True)
```

```python
print(df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

```python
print(df.describe())
```

```python
target_column = 'Gender'

if df[target_column].dtype == 'object':
    print("Це класифікаційна задача")
else:
    print("Це задача регресії")
```

```python
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[['Age', 'Height', 'Weight', 'Medu', 'Fedu']])  

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)
kmeans_silhouette = silhouette_score(X_scaled, kmeans_labels)

dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)
dbscan_silhouette = silhouette_score(X_scaled, dbscan_labels) if len(set(dbscan_labels)) > 1 else -1

agglo = AgglomerativeClustering(n_clusters=3)
agglo_labels = agglo.fit_predict(X_scaled)
agglo_silhouette = silhouette_score(X_scaled, agglo_labels)

print(f"K-Means Silhouette Score: {kmeans_silhouette}")
print(f"DBSCAN Silhouette Score: {dbscan_silhouette}")
print(f"Agglomerative Clustering Silhouette Score: {agglo_silhouette}")
```

```python
inertia = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(range(2, 11), inertia, marker='o')
plt.title('Метод ліктя для K-Means')
plt.xlabel('Кількість кластерів')
plt.ylabel('Inertia')
plt.show()
```

```python
print(f"K-Means Silhouette Score: {kmeans_silhouette}")
print(f"DBSCAN Silhouette Score: {dbscan_silhouette}")
print(f"Agglomerative Clustering Silhouette Score: {agglo_silhouette}")
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt

kmeans = KMeans(n_clusters=4, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=kmeans_labels, palette='Set1')
plt.title('Візуалізація на основі кластерів')
plt.show()
```

Метод ліктя вказує на оптимальну кількість кластерів між 4 і 5, оскільки після цієї точки інерція зменшується менш різко. Візуалізація з 4 кластерами показує добре розділені групи, де кожен кластер чітко відокремлюється в PCA-просторі.
